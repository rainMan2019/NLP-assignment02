```python
import random
url_host = 'http://www.bjsubway.com/station/xltcx/'  #北京地铁网站
wb_data = requests.get(url_host)
wb_data.encoding = 'gb18030' 
soup=BeautifulSoup(wb_data.text,'lxml')
bj_station_route={}
for i in range(200,403):
    station=soup.select('body > div > div.content > div.sub_map > div.line_content > div:nth-child({})'.format(i))[0].text
    if re.search(r'线',station)!=None:
        station = None
    station_next=soup.select('body > div > div.content > div.sub_map > div.line_content > div:nth-child({})'.format(i+1))[0].text
    if re.search(r'线',station_next)!=None:
        station_next=None
    station_pre=soup.select('body > div > div.content > div.sub_map > div.line_content > div:nth-child({})'.format(i-1))[0].text
    if re.search(r'线',station_pre)!=None:
        station_pre=None
    if station!=None:
        if station in bj_station_route.keys(): 
            if station_pre!=None:
                bj_station_route[station].append(station_pre)
            if station_next!=None:
                bj_station_route[station].append(station_next)
        else:
            bj_station_route[station]=[]
            if station_pre!=None:
                bj_station_route[station].append(station_pre)
            if station_next!=None:
                bj_station_route[station].append(station_next)
        print(bj_station_route[station])
```
